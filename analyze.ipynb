```python
#یک جدول کوچک خلاصه از روی این جدول می سازیم 
table_for_best_selling_product = cleaned_wood_sale.loc[:,['sale_date', 'Item_ID', 'item_name', 'quantity', 'total_sale', 'net_price']]
table_for_best_selling_product.head(3)
```

```python
#جدول تامین را هم در نظر می گیریم و از آن یک جدول خلاصه کوچک استخراج می کنیم
table_for_best_selling_product_from_suppliers = cleaned_wood_purchase[['item_ID', 'item_name', 'quantity', 'total_price', 'net_price']]
#مثبت کردن ستون های منفی
table_for_best_selling_product_from_suppliers.loc[:,'total_price'] = table_for_best_selling_product_from_suppliers['total_price'].abs()
table_for_best_selling_product_from_suppliers.loc[:,'net_price'] = table_for_best_selling_product_from_suppliers['net_price'].abs()
table_for_best_selling_product_from_suppliers.head(3)
```

```python
#حال هر دو جدول را بر اساس نام کالا گروپ بای می کنیم
#دیتاست فروش - مشتریان
table_for_best_selling_product = table_for_best_selling_product.groupby('item_name').agg(
    quantity = ('quantity', 'sum'),
    total_sale = ('total_sale', 'sum'),
    net_price = ('net_price', 'sum')
).reset_index()
table_for_best_selling_product.head()
```

```python
#دیتاست خرید - تامین کنندگان
table_for_best_selling_product_from_suppliers = table_for_best_selling_product_from_suppliers.groupby('item_name').agg(
    quantity = ('quantity', 'sum'),
    total_price = ('total_price', 'sum'),
    net_price = ('net_price', 'sum')
).reset_index()
table_for_best_selling_product_from_suppliers.head()
```

```python
#حالا این دو جدول را بر اساس نام کالاها مرج می کنیم
Comprehensive_table_for_best_selling_product = pd.merge(table_for_best_selling_product_from_suppliers,
                                                        table_for_best_selling_product,
                                                        on = 'item_name',
                                                        suffixes = ('_Supplied','_Sold'),
                                                        how = 'left')

Comprehensive_table_for_best_selling_product.head()
```

```python
#حال با اختلاف مبلغ خالص فروش به مبلغ خالص خرید سود خالص بدست می آید
#پس این ستون را ایجاد می کنیم
Comprehensive_table_for_best_selling_product['NET_Profit'] = Comprehensive_table_for_best_selling_product['net_price_Sold'] - Comprehensive_table_for_best_selling_product['net_price_Supplied']
Comprehensive_table_for_best_selling_product.head()
```

```python
#حال بر اساس تعداد فروش مرتب سازی می کنیم
Comprehensive_table_for_best_selling_product = Comprehensive_table_for_best_selling_product.sort_values('quantity_Sold', ascending=False)
Comprehensive_table_for_best_selling_product.head(3)
```

```python
#سود هر واحد از کالاها را جداگانه حساب می کنیم
Comprehensive_table_for_best_selling_product['Profit_Per_Unit'] = (
    (Comprehensive_table_for_best_selling_product['NET_Profit'] / Comprehensive_table_for_best_selling_product['quantity_Sold'])
    .round(2)
)
Comprehensive_table_for_best_selling_product.head(5)
```

```python
#ابتدا هر دو جدول تامین کنندگان و مشتریان را فراخوانی می کنیم
#ستون های مورد نیاز از دیتاست مشتریان
Summary_table_of_customers = cleaned_wood_sale[['item_name','quantity']]
Summary_table_of_customers.head()
```

```python
#دیتا ست فروش
# حال از دستور گروپ بای استفاده می کنیم و برای هر محصول جمع مقادیر ستون تعداد را محاسبه می کنیم
Summary_table_of_groupped_customers = Summary_table_of_customers.groupby(['item_name']).agg(
    Amount_of_product = ('quantity', 'sum')
).reset_index()

Summary_table_of_groupped_customers.head()
```

```python
#همین کار را برای دیتا ست خرید انجام می دهیم
Summary_table_of_groupped_suppliers = Summary_table_of_suppliers.groupby(['item_name']).agg(
   Amount_of_product = ('quantity', 'sum')
).reset_index() 
Summary_table_of_groupped_suppliers.head()
```

```python
#حال از دستور مرج استفاده می کنیم و جدول دیگری تولید می کنیم به شرح زیر
first_solution_result = pd.merge(Summary_table_of_groupped_suppliers,
                                 Summary_table_of_groupped_customers,
                                 on = 'item_name',
                                 suffixes = ('_Supplied','_Sold'),
                                 how = 'left')
first_solution_result.head(5)
```

```python
# با توجه به اینکه ما لفت مرج انجام دادیم لذا ممکن است بعضی از کالاهایی که تامین کرده ایم هنوز به فروش نرفته باشند
#پس مقدار آنها در ستون وسطی برابر با نان خواهد بود
# که کار محاسبه را با خطا مواجه خواهد ساخت
#ابتدا یک دستور می نویسیم تا هر جا در ستون وسطی نان دید آنرا تبدیل به صفر کند.
first_solution_result['Amount_of_product_Sold'] = first_solution_result['Amount_of_product_Sold'].mask(first_solution_result['Amount_of_product_Sold'] < 0).fillna(0)

first_solution_result.sort_values('Amount_of_product_Sold', ascending=False)
```

```python
#یک ستون به این جدول اضافه می کنیم که اختلاف دو ستون دیگر را به نام موجودی انبار نمایش دهد
first_solution_result['quantity_difference'] = first_solution_result['Amount_of_product_Supplied']-first_solution_result['Amount_of_product_Sold']
first_solution_result.head()
```

```python
#با توجه به اینکه موجودی انبار منطقا نمی تواند منفی باشد
#پس احتمالا قبل از شروع سال در بعضی از ردیف ها موجودی انبار داشته ایم
#لذا اعداد منفی ستون موجودی انبار را صفر می کنیم

first_solution_result['quantity_difference'] = first_solution_result['quantity_difference'].mask(first_solution_result['quantity_difference'] < 0).fillna(0)
first_solution_result.head()
```

```python
#حال جمع کل موجودی انبار را محاسبه می کنیم
Total_warehouse_products = round(first_solution_result.quantity_difference.sum(),2)
print("Total warehouse Products = ",Total_warehouse_products)
```

```python
#ستون جدید اضافه می کنیم که درصد موجودی هر کدام از کالاها نسبت به  موجودی کل انبار را محاسبه کند
first_solution_result['inventory_percentage'] = round(
    (100*(first_solution_result["quantity_difference"]/Total_warehouse_products)), 2)
first_solution_result.head()
```
